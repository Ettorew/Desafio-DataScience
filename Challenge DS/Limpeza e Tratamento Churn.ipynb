### Outliers e Duplicatas:

```python
df.describe()
```

```python
def calcula_outliers(dataframe, coluna):
    Q1 = dataframe[coluna].quantile(.25)
    Q3 = dataframe[coluna].quantile(.75)
    IQR = Q3 - Q1
    valor_min = Q1 - IQR*1.5
    valor_max = Q3 + IQR*1.5
    
    outlier = (dataframe[coluna] < valor_min) | (dataframe[coluna] > valor_max)
    return outlier
```

```python
outliers_index = calcula_outliers(df, 'customer.tenure')
outliers_index
```

```python
df[outliers_index]['customer.tenure']
```

```python
outliers_index = calcula_outliers(df, 'account.Charges.Monthly')
df[outliers_index]['account.Charges.Monthly']
```

Considerando que todos os casos não há Churn, o desvio-padrão consideravelmente alto, porém um max que não foge, já um min que foge 69x dele, vemos que ele tem uma margem de 4,1% de outliers, ao retirá-los:

Percebe-se que também a quantia de contratos de dois anos que não possui o Churn com e sem os outliers é dominante, sendo de cerca de 97%. Também pode-se notar que houve um certo "boom" no período entre 3 e 2 anos antes da coleta dos dados no contrato de dois anos (3 anos antes da coleta sendo o início de quando os dados começaram a entrar), e houve uma repentina queda da assinatura desse contrato. Pela ínfima quantidade de dados vazios que estou trabalhando (11 no total), e apesar da alta chance de eles serem desse período de 2/3 anos antes da coleta dos dados, não há nada que possa confirmar isso, eu optei por tirá-los do dataframe, e irei trabalhar sem eles, porém com os outliers, que apesar de fugirem muito do desvio-padrão, dado ao contexto, são dados válidos.
